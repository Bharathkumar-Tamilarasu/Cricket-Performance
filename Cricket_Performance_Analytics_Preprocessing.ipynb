```python
### Adding all the matchdictionaries values (i.e List of Dictionaries)
### Batting summary will have each batting record of a single match
### i.e For Data[0] match have 19 innings, Data[2] match have 19 innings....
### Data[44] innings we have 17 innings (because only 17 players played batting inning in that match)
### ... In pakistan 10 players and in England 7 players


batting_summary_list = []

for i in range(len(data)):
  batting_summary_list.extend(list(data[i].values())[0])

print(len(batting_summary_list))
```

```python
### List of Dictionaries

batting_summary_list[0:2]
```

```python
type(batting_summary_list[0])
```

```python
### Creating a DataFrame from Batting Summary list

df_batting = pd.DataFrame(batting_summary_list)
df_batting.head(10)
```

```python
df_batting.tail()
```

```python
df_batting['out/notout'] = df_batting['dismissal'].apply(lambda x:'notout' if x == '' else 'out')
```

```python
df_batting.drop(['dismissal'],axis=1,inplace=True)
```

```python
df_batting.tail()
```

```python
### Transforming Improper Names to Proper Names

### Retreiving values where batsman name is not proper

df_batting[df_batting['batsmanName'].apply(lambda x: True if re.match("^[\.\-\'\s\(\)A-Za-z]+$",x) == None else False)].batsmanName.unique()
```

```python
### Replacing improper characters in batsman names (i.e Replacing the characters which are not '.-A-Za-z() ')
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x:re.sub('[^\.\-\'\s\(\)A-Za-z]','',x))
```

```python
df_batting.head(10)
```

```python
df_batting[df_batting['batsmanName'].apply(lambda x: True if re.match("^[\.\-\'\s\(\)A-Za-z]$",x) else False)].batsmanName.unique()
```

```python
### From the above result we can understand that there is no improper names in 'batsmanName' columns
```

```python
### Creating Match ID column for this data frame
```

```python
df_batting.head()
```

```python
### Creating a dictionary of MatchIDs from df_match

match_id_dict = {}
for index,row in df_match.iterrows():
  k1 = row['team1'] + ' Vs ' + row['team2']
  k2 = row['team2'] + ' Vs ' + row['team1']
  match_id_dict[k1] = row['matchId']
  match_id_dict[k2] = row['matchId']
```

```python
### Applying match id for each row with respect to its match column

df_batting['matchId'] = df_batting['match'].apply(lambda x:match_id_dict[x])
```

```python
df_batting.head(5)
```

```python
df_batting.tail(5)
```

```python
### Creating a Batting summary csv

df_batting.to_csv('/content/drive/MyDrive/Colab Notebooks/Self Projects/Cricket Performance Analytics/csv files/batting_summary.csv',index=False)
```

## **iv) Player Summary**

```python
with open('/content/drive/MyDrive/Colab Notebooks/Self Projects/Cricket Performance Analytics/json files/t20_wc_player_info.json') as f:
  data = json.load(f)
```

```python
type(data) ### Data is a type of List of Dictionaries
```

```python
df_players = pd.DataFrame(data)
df_players.head()
```

```python
df_players.info()
```

```python
df_players.isnull().sum()
```

```python
df_players.columns
```

```python
df_players[df_players['name'].apply(lambda x:False if re.match('^[\.\-\'\s\(\)A-Za-z]+$',x) else True)]
```

```python
print(list(df_players.columns).remove('description'))
```

```python
### Replacing improper characters in batsman names (i.e Replacing the characters which are not '.'-A-Za-z() ')

for i in df_players.columns:
  if i != 'description':
    df_players[i] = df_players[i].apply(lambda x:re.sub('[^\.\-\'\s\(\)A-Za-z]','',x))
```

```python
### Player names After replacing

df_players[80:86]
```

```python
### Checking if we have any improper names in all columns

for i in df_players.columns:
  if i != 'description':
    print(i,' -> ',df_players[df_players[i].apply(lambda x:False if re.match('^[\.\-\'\s\(\)A-Za-z]+$',x) else True)][i].unique())
```

```python
df_players.to_csv('/content/drive/MyDrive/Colab Notebooks/Self Projects/Cricket Performance Analytics/csv files/players_summary.csv',index=False)
```
